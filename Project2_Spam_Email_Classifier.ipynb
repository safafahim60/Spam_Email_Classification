**Converting Categorical Feature into Numeric**

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
dataset['Email No.'] = le.fit_transform(dataset['Email No.'])
dataset.head()
```

**Separate feature variable and target variable**

```python
Y= dataset["Prediction"]
X= dataset.drop(columns=["Email No.","Prediction"])
```

```python
# Import the StandardScaler class
from sklearn.preprocessing import StandardScaler

# Create an instance of the StandardScaler class
scaler = StandardScaler()

# Fit the scaler to the data in X
X_scaled = scaler.fit_transform(X)
```

# New Section

```python
spam_count = sum(dataset['spam'])
non_spam_count = len(dataset) - spam_count
total_emails = len(dataset)
print("\nSpam emails:", spam_count)
print("Non-spam emails:", non_spam_count)
print("Total emails:", total_emails)

categories = ['Spam', 'Non-spam']
counts = [spam_count, non_spam_count]

plt.bar(categories, counts, color=['red', 'green'])
plt.xlabel('Email Type')
plt.ylabel('Count')
plt.title('Count of Spam and Non-spam Emails')
plt.show()
```

# Model Selection

**Models**

```python
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
Models ={"LR":LogisticRegression(),
  "DT": DecisionTreeClassifier(),
  "RF": RandomForestClassifier(),
  "GB": GradientBoostingClassifier()}
```

# Model Training

```python
from sklearn.model_selection import train_test_split
for name, clf in Models.items():
    print(f"Train and evaluate { name}")
    X_train,X_test,Y_train,Y_test=train_test_split(X_scaled,Y, test_size=0.2, random_state=42)
    clf.fit(X_train,Y_train)
```

```python
models = []
accuracies = []
precisions = []
recalls = []
#f1_score = []
from sklearn.metrics import accuracy_score,f1_score,precision_score,recall_score
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report,confusion_matrix
for model,clf in Models.items():
    print(f"Train and evaluate { model}")
    X_train,X_test,Y_train,Y_test=train_test_split(X_scaled,Y, test_size=0.2, random_state=42)
    clf.fit(X_train,Y_train)
    predictions = clf.predict(X_test)

    #Performance Matrices
    accuracy= accuracy_score(Y_test,predictions)
    precision= precision_score(Y_test,predictions)
    recall= recall_score(Y_test,predictions)

    #Append metrices -> list
    models.append(model)
    accuracies.append(accuracy)
    precisions.append(precision)
    recalls.append(recall)

    #f1_score= f1_score(Y_test,predictions,pos_label=1)
    print (f"Accuracy{accuracy:2f}")
    print (f"precision{precision:2f}")
    print (f"recall{recall:2f}")
   #print (f"f1_score{f1_score:2f}")
    print()
    print (f'clasification report:', classification_report(Y_test,predictions))


    print (f'\n confussion matrix:',confusion_matrix(Y_test, predictions))
    cm = confusion_matrix(Y_test, predictions)
    plt.figure(figsize=(10,7))
    sns.heatmap(cm, annot=True,cmap='Blues', fmt='g')
    plt.xlabel('Predicted')
    plt.ylabel('Truth')
```

**Model Evaluation with cross-validation**

```python
from sklearn.model_selection import cross_val_score
for model,clf in Models.items():
  print(f"Evaluate model { model} with cross validation")
  #score_list = ["roc_auc", "accuracy", "f1", "precision", "recall"]
  cv_score= cross_val_score(clf,X_scaled,Y, cv=5, scoring= "accuracy")
  print(f"Mean cross validation acc{cv_score.mean():2f}")
# score_list = ["roc_auc", "accuracy", "f1", "precision", "recall"]
# plot_cv_score(X = X_scaled, y = Y, models_list = clf, cv = 4, scoring_list = score_list, refit = True)
```
